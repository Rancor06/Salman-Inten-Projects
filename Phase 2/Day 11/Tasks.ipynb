```python
#TASK 1: Upload the data set
from google.colab import files
import io

# Upload file from local drive
uploaded = files.upload()
```

```python
# TASK 2: Import Required Libraries
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

print("pandas:", pd.__version__)
print("numpy:", np.__version__)
print("sklearn:", sklearn.__version__)
```

```python
# TASK 3: Load the Dataset
df = pd.read_csv(io.BytesIO(uploaded['student_performance.csv']))
print("First 5 values of the Dataset:\n",df.head())
print("\nLast 5 values of the Dataset:\n",df.tail())
print("\nShape of the Dataset: ",df.shape)
print("\nDataset Information:")
df.info()
```

```python
# TASK 4: Select Features and Targets
X = df.drop(columns=['student_id', 'grade','total_score'])
Y = df['grade']
print(X)
print(Y)
```

```python
# TASK 5: Train - Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
print("Training Sample Shape: ",X_train.shape)
print("Testing Sample Shape: ",X_test.shape)
```

```python
# TASK 6: Train the Model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
print("\nDecision Tree Classifier trained successfully.")
```

```python
# TASK 7: Make Predictions
y_pred = model.predict(X_test)
print("\nPredictions on test data:")
print("First 10 predicted values:", y_pred[:10])
print("First 10 actual values:", y_test[:10].values)
```

```python
# TASK 8: Predict New Data
new_students = pd.DataFrame({
    'weekly_self_study_hours': [20.0, 10.5, 3.0],
    'attendance_percentage':   [92.0, 75.0, 45.0],
    'class_participation':     [4.5, 2.8, 0.9]
})
predictions = model.predict(new_students)
print(predictions)
```
